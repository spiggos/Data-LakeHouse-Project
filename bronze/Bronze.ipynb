#Bronze Layer

## Set up Ingestion Configuration

```python
BASE_URL = '/Volumes/workspace/bronze/raw_sources/'

INGESTION_CONFIG = [
{
        "source": "crm",
        "path": F"{BASE_URL}/source_crm/cust_info.csv",
        "table": "crm_cust_info"
},
{
        "source": "crm",
        "path": F"{BASE_URL}/source_crm/prd_info.csv",
        "table": "crm_prd_info"
},
{
        "source": "crm",
        "path": F"{BASE_URL}/source_crm/sales_details.csv",
        "table": "crm_sales_details"
},
{
        "source": "erp",
        "path": F"{BASE_URL}/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
},
{
        "source": "erp",
        "path": F"{BASE_URL}/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
},
{
        "source": "erp",
        "path": F"{BASE_URL}/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
}

]
```

## Ingest data from Source to Bronze Table

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze.{item['table']}")

    df = (spark.read.format("csv")
    .option("header", True)
    .option("inferSchema", True)
    .load(item['path'])
    )
    df.write.format("delta").mode("overwrite").saveAsTable(f"workspace.bronze.{item['table']}")
```
